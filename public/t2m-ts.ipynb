```python
import plotly.graph_objects as go
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("era5_daily_series_2t_global_1940-2023.csv")
df
```

```python
df[df['date'].str.contains("1940")==True]
```

```python
fig = go.Figure()

for year in range(1940, 2024):
    data = df[df['date'].str.contains(str(year))==True].temp.values
    fig.add_trace(go.Scatter(x=list(range(len(data))), y=data, name=str(year)))
    

fig.update_layout(autosize=True)
```
